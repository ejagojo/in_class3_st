```python
from PIL import Image, ImageEnhance
from IPython.display import display, Markdown

# Load images
try:
    word_matrix = Image.open("word_matrix.png").convert("RGB")
    mask = Image.open("mask.png").convert("L")  # Convert mask to grayscale
except FileNotFoundError:
    display(Markdown("**Error:** One or both image files are missing. Ensure 'word_matrix.png' and 'mask.png' are in the working directory."))
    raise

# Ensure both images have the same size before merging
mask = mask.resize(word_matrix.size)

# Enhance the mask contrast for better visibility
mask = ImageEnhance.Contrast(mask).enhance(2.5)  # Increase contrast

# Apply transparency (keeping only the masked parts visible)
binary_mask = mask.point(lambda p: 255 if p > 128 else 0)  # Convert to binary mask

# Create the final overlaid image
blended = Image.composite(word_matrix, Image.new("RGB", word_matrix.size, "white"), binary_mask)
display(blended)
display(Markdown("Great work with images"))
display(Markdown("You are the best"))
blended.show()
```
