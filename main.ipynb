```python
from PIL import Image, ImageEnhance
from IPython.display import display, Markdown

try:
    # 1. Load images
    word_matrix = Image.open("word_matrix.png").convert("RGB")
    mask = Image.open("mask.png").convert("L")  # Grayscale for masking
except FileNotFoundError:
    display(Markdown("**Error:** One or both image files are missing. Ensure 'word_matrix.png' and 'mask.png' are in the working directory."))
    raise

# 2. Ensure both images have the same size
if mask.size != word_matrix.size:
    mask = mask.resize(word_matrix.size)
    
blended = Image.composite(word_matrix, mask, mask)
display(blended) # Display Blended Image

# Display the hidden messages in the Mark down
display(Markdown("GREAT WORK WITH IMAGES")) 
display(Markdown("YOU ARE THE BEST"))

blended.show()

```
